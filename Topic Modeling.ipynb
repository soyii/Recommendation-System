### Topic modeling for movie reviews

data : naver movie reviews

```python
import tomotopy as tp
```

```python
#k = number of topic
model = tp.LDAModel(k=20, alpha=0.1, eta=0.01, min_cf=5)
```

```python
for i, line in enumerate(open('ratings.txt', encoding='utf-8')):
    model.add_doc(line.strip().split())
    if i % 10 == 0: print('Document #{} has been loaded'.format(i))
```

```python
model.train(0)
print('Total docs:', len(model.docs))
print('Total words:', model.num_words)
print('Vocab size:', model.num_vocabs)
```

```python
for i in range(200):
    print('Iteration {}\tLL per word {}'.format(i, model.ll_per_word))
    model.train(1)
```

```python
for i in range(model.k):
    res = model.get_topic_words(i, top_n=10)
    print('Topic #{}'.format(i), end='\t')
    print(', '.join(w for w, p in res))
```

```python
#preprocessing korean
from kiwipiepy import Kiwi

kiwi = Kiwi()
kiwi.prepare()
stopwords = set(["영화", "너무", "이다", "이런", "하","이", "거", "하다", "나다", "다"])
def tokenize(sent):
    res, score = kiwi.analyze(sent)[0]
    return [word + ('다' if tag.startswith('V') else '')
           for word, tag, _, _ in res
           if not tag.startswith('E') and not tag.startswith('J') and not tag.startswith('S')]
```

```python
model = tp.LDAModel(k=20, alpha=0.1, eta=0.01, min_cf=5)
for i, line in enumerate(open('ratings.txt', encoding='utf-8')):
    model.add_doc(tokenize(line))
    if i % 10 == 0: print('Document #{} has been loaded'.format(i))
        
for i in range(200):
    print('Iteration {}\tLL per word: {}'.format(i, model.ll_per_word))
    model.train(1)
```

```python
for i in range(model.k):
    res = model.get_topic_words(i, top_n=10)
    print('Topic #{}'.format(i), end='\t')
    print(', '.join(w for w, p in res))
```
